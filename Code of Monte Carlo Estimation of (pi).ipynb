# Task 2: Monte Carlo Estimation of π (pi) 

Part 1: Monte Carlo Simulation 


# A. Design the Monte Carlo Simulation: 

```python
import random

def estimate_pi(num_points):
    inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(0, 1)  # Generate a random x-coordinate between 0 and 1
        y = random.uniform(0, 1)  # Generate a random y-coordinate between 0 and 1

        # Check if the point is inside the quarter-circle (x^2 + y^2 <= 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1

    # Estimate pi using the Monte Carlo method
    estimated_pi = 4 * (inside_circle / num_points)
    return estimated_pi

if __name__ == "__main__":
    num_points = int(input("Enter the number of random points to generate: "))
    estimated_pi = estimate_pi(num_points)
    print(f"Estimated value of pi using {num_points} points: {estimated_pi}")

```

# C. Determine Points Inside the Quarter-Circle: 
● Implement logic to determine which generated points fall inside the quarter-circle with a radius of 1 and centered at the origin (0, 0). 
● Track the number of points that fall inside the quarter-circle. 


```python
import random
import matplotlib.pyplot as plt

def generate_random_points(N):
    points = []
    for _ in range(N):
        x = random.uniform(0, 1)  # Generate a random x-coordinate between 0 and 1
        y = random.uniform(0, 1)  # Generate a random y-coordinate between 0 and 1
        points.append((x, y))
    return points

def points_inside_quarter_circle(points):
    inside_count = 0
    for x, y in points:
        if x**2 + y**2 <= 1:  # Check if the point is inside the quarter-circle
            inside_count += 1
    return inside_count

# Specify the number of random points to generate
N = 1000

# Generate random (x, y) coordinates within the unit square
random_points = generate_random_points(N)

# Determine and track the number of points inside the quarter-circle
inside_count = points_inside_quarter_circle(random_points)

# Visualize the points and quarter-circle
x_inside = []
y_inside = []
x_outside = []
y_outside = []

for x, y in random_points:
    if x**2 + y**2 <= 1:
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

# Create a scatter plot
plt.scatter(x_inside, y_inside, color='blue', label='Inside Quarter-Circle')
plt.scatter(x_outside, y_outside, color='red', label='Outside Quarter-Circle')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Monte Carlo Simulation - Points Inside Quarter-Circle')
plt.legend()

# Draw the quarter-circle
circle = plt.Circle((0, 0), 1, color='green', fill=False, linestyle='dashed', linewidth=2)
plt.gca().add_patch(circle)

# Display the plot
plt.axis('equal')
plt.show()

# Print the result
print(f"Total points inside the quarter-circle: {inside_count}/{N}")
```

# Part 2: Estimating π (pi) 
A. Calculate π Estimate: 
● Calculate an estimate of π using the following formula: 

● π (pi) ≈ (Number of Points Inside the Quarter-Circle / Total Number of Points) * 4 


```python
import random

# Number of random points to generate
N_total = 1000000  # You can adjust this number for more accuracy
N_in = 0

# Generate random points and count those inside the quarter-circle
for _ in range(N_total):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    distance = x**2 + y**2  # Calculate the square of the distance from the origin
    if distance <= 1:
        N_in += 1

# Calculate the estimate of π
pi_estimate = (N_in / N_total) * 4
print("Estimated value of π:", pi_estimate)
```
