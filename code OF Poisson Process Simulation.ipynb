# Task 3: Poisson Process Simulation and Analysis 
# Part 1: Simulation 
# A. Design the Poisson Process Simulation: 
● Create a Python program that simulates events occurring in a Poisson process. ● Specify the average rate (λ) of events per unit time as a parameter. 

```python
import numpy as np
import matplotlib.pyplot as plt

def poisson_process_simulation(rate, time_period):
    time = 0
    event_times = []
    
    while time < time_period:
        # Generate time until the next event using an exponential distribution
        time_to_next_event = np.random.exponential(1 / rate)
        
        # Update the current time
        time += time_to_next_event
        
        # Add the event time to the list
        event_times.append(time)
    
    return event_times

# Parameters
average_rate = 5  # λ, average rate of events per unit time
simulation_time = 10  # Total simulation time

# Simulate the Poisson process
event_times = poisson_process_simulation(average_rate, simulation_time)

# Plot the event times
plt.figure(figsize=(10, 5))
plt.step(event_times, range(len(event_times)), where='post', label='Event Times')
plt.xlabel('Time')
plt.ylabel('Number of Events')
plt.title('Poisson Process Simulation')
plt.legend()
plt.grid()
plt.show()
```

# B. Generate Poisson Events: 
● Implement the simulation logic to generate Poisson events over a specified time interval (e.g., one hour). 
● Record the timestamps of each event occurrence. 

```python
import numpy as np

def poisson_events_simulation(rate, time_interval):
    event_times = []
    time = 0

    while time < time_interval:
        # Generate time until the next event using an exponential distribution
        time_to_next_event = np.random.exponential(1 / rate)
        
        # Check if the next event time exceeds the time interval
        if time + time_to_next_event <= time_interval:
            # Update the current time
            time += time_to_next_event
            
            # Record the event time
            event_times.append(time)
        else:
            break

    return event_times

# Parameters
average_rate = 5  # λ, average rate of events per unit time
simulation_time = 60  # Total simulation time in minutes (e.g., one hour)

# Convert the simulation time to seconds
simulation_time_seconds = simulation_time * 60

# Simulate the Poisson events
event_times = poisson_events_simulation(average_rate, simulation_time_seconds)

# Print the event timestamps
for i, event_time in enumerate(event_times):
    print(f"Event {i+1} timestamp: {event_time} seconds")

# Calculate the total number of events
total_events = len(event_times)
print(f"Total number of events: {total_events}")
```

# Part 2: Data Analysis 
# A. Calculate Event Statistics: 
● After running the simulation, calculate the following statistics based on the generated event data: 
○ Total number of events observed. 
○ Inter-arrival times between consecutive events. 
○ The average and standard deviation of inter-arrival times. 

```python
import numpy as np

def poisson_events_simulation(rate, time_interval):
    event_times = []
    time = 0

    while time < time_interval:
        # Generate time until the next event using an exponential distribution
        time_to_next_event = np.random.exponential(1 / rate)
        
        # Check if the next event time exceeds the time interval
        if time + time_to_next_event <= time_interval:
            # Update the current time
            time += time_to_next_event
            
            # Record the event time
            event_times.append(time)
        else:
            break

    return event_times

# Parameters
average_rate = 5  # λ, average rate of events per unit time
simulation_time = 60  # Total simulation time in minutes (e.g., one hour)

# Convert the simulation time to seconds
simulation_time_seconds = simulation_time * 60

# Simulate the Poisson events
event_times = poisson_events_simulation(average_rate, simulation_time_seconds)

# Calculate the total number of events
total_events = len(event_times)
print(f"Total number of events: {total_events}")

# Calculate inter-arrival times
inter_arrival_times = np.diff(event_times)

# Calculate the average and standard deviation of inter-arrival times
average_inter_arrival_time = np.mean(inter_arrival_times)
std_deviation_inter_arrival_time = np.std(inter_arrival_times)

print(f"Average inter-arrival time: {average_inter_arrival_time} seconds")
print(f"Standard deviation of inter-arrival time: {std_deviation_inter_arrival_time} seconds")
```
